```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment_in_India.csv")

df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()

df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.isnull().sum()
```

```python
df['Estimated Unemployment Rate (%)'].mean()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)

plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=df)

plt.xticks(rotation=90)
plt.title("Unemployment Rate by Region")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Region', y='Estimated Unemployment Rate (%)', data=df)

plt.xticks(rotation=90)
plt.title("Unemployment Distribution During COVID-19")
plt.show()
```
